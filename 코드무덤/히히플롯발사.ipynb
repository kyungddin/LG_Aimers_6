```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import prince

from sklearn.preprocessing import  OrdinalEncoder
from sklearn.ensemble import ExtraTreesClassifier

train = pd.read_csv('./train.csv').drop(columns=['ID'])
test = pd.read_csv('./test.csv').drop(columns=['ID'])
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager

# Windows에서 사용할 수 있는 한글 폰트 경로 설정 (예: Malgun Gothic)
font_path = "C:\\Windows\\Fonts\\malgun.ttf"  # 윈도우 기본 폰트인 Malgun Gothic을 사용
font_prop = font_manager.FontProperties(fname=font_path)

# 폰트 설정
plt.rcParams['font.family'] = font_prop.get_name()

# 이제 그래프를 그릴 때 한글이 깨지지 않습니다.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 여러 feature들 간의 pairwise 관계 시각화
sns.pairplot(train[['총 생성 배아 수', '미세주입에서 생성된 배아 수', '이식된 배아 수',
                   '미세주입 배아 이식 수', '저장된 배아 수',
                   '미세주입 후 저장된 배아 수', '해동된 배아 수']])
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 플롯 크기 설정 (가로 12인치, 세로 8인치)
plt.figure(figsize=(20, 20))

# 여러 feature의 분포 비교
sns.boxplot(data=train[['총 생성 배아 수', '미세주입에서 생성된 배아 수', '이식된 배아 수',
                   '미세주입 배아 이식 수', '저장된 배아 수',
                   '미세주입 후 저장된 배아 수', '해동된 배아 수']])

plt.title('Boxplot of Features')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 선형 회귀선 추가한 산점도
sns.regplot(x='저장된 배아 수', y='총 생성 배아 수', data=train)
plt.title('Regression Plot: 저장된 배아 수 vs 총 생성 배아 수')
plt.xlabel('저장된 배아 수')
plt.ylabel('총 생성 배아 수')
plt.show()
```
